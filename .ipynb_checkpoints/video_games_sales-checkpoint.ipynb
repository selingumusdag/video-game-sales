```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(40,30))
y=len(games.columns)
games2=games[games["Global_Sales"]<10]
for i in range(6,y):
    plt.subplot(5,3,i-2)
    plt.hist(games2[games2.columns[i]],color = "blue",bins=50)
    plt.title(games2.columns[i])
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import zscore
my_list=[]
z_scores = zscore(games["EU_Sales"])
for threshold in np.arange(1,5,0.5):
    #print("Eik deeri: {}".format(threshold))
    #print("Aykr deerlerin says: {}".format(len((np.where(z_scores > threshold)[0]))))
    my_list.append((threshold,len((np.where(z_scores > threshold)[0]))))


df_treshold=pd.DataFrame(my_list,columns=['Treshold','Count of EU_Sale'])
plt.plot(df_treshold['Treshold'],df_treshold['Count of EU_Sale'])
```

```python
def categorized_sales(col,liste):
    if col>liste[1]:
        return 1
    elif col>=liste[0]:
        return 2
    else:
        return 3
    
```

```python
a=max(games["NA_Sales"])*0.70
print('{0:.2f}'.format(a))
```

```python
games['NA_SaleGroup']=games['NA_Sales'].apply(lambda x: categorized_sales(x,[1,20]))
games['EU_SaleGroup']=games['EU_Sales'].apply(lambda x: categorized_sales(x,[1,10]))
games['JP_SaleGroup']=games['JP_Sales'].apply(lambda x: categorized_sales(x,[1,5]))
games['Other_SaleGroup']=games['Other_Sales'].apply(lambda x: categorized_sales(x,[1,4]))
display(games.head())
```
